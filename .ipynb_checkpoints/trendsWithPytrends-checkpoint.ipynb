```python
kw_list = ['pizza']
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
pytrend.build_payload(kw_list)
# Interest by Region
df = pytrend.interest_by_region()
df.head(10)
#specify how many you want with .head
# df.reset_index().plot(x='geoName', y='Taylor Swift', figsize=(120, 10), kind ='bar')
```

```python
pytrend.interest_over_time()
df.head(10)
```

```python
df = pytrend.get_historical_interest(kw_list, year_start=2020, month_start=1, day_start=1, hour_start=0, year_end=2020, month_end=2, day_end=1, hour_end=0, cat=0, geo='', gprop='', sleep=5)
df.head(10)
```

```python
x= df['pizza']
for i in x:
    print(x[i])
```

```python
df = pytrend.trending_searches(pn='united_states')
df.head()
```

```python
df = pytrend.today_searches(pn='US')
df.head()
```

```python
# Get Google Top Charts
df = pytrend.top_charts(2020, hl='en-US', tz=300, geo='GLOBAL')
df.head(10)
```
